```python
frames = [df2018,df2017,df2016, df2015, df2014]
master_df = pd.concat(frames, sort = False)
master_df
```

```python
master_df = master_df.dropna()
master_df
```

```python
starts = master_df['starttime'].str.split(" ",n =1, expand = True)
master_df['startdate'] = starts[0]
master_df['startingtime'] = starts[1]

ends = master_df['stoptime'].str.split(" ",n=1, expand=True)
master_df['enddate'] = ends[0]
master_df['endtime'] = ends[1]



master_df
```

```python
master_df.drop(columns = ['starttime'], inplace = True)
master_df.drop(columns = ['stoptime'], inplace = True)
```
